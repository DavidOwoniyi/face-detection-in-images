# Face Detection in images

### This simple program detects faces in images using Python,
### OpenCV, numpy and matplotlib. Steps are as follows.

<img src="sample.PNG" alt="Face Detection" width=500px/>

### Import libraries

```python
import cv2 as cv
import numpy as np
import requests
import matplotlib.pyplot as plt
```

### URL of the hosted image

```python
url = "https://lh3.googleusercontent.com/d/1SazmxZvko0PO8DiDQtodRHHoMl50UJyv"
```

### Download the image

```python
response = requests.get(url)
img_array = np.array(bytearray(response.content), dtype=np.uint8)
```

### Convert image to OpenCV format

```python
img = cv.imdecode(img_array, -1)
```

### Convert image to grayscale

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

### Load the Classifier

```python
face_classifier = cv.CascadeClassifier(
    cv.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

### Perform face detection

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=7, minSize=(40, 40)
)
```

### Draw a bounding circle

```python
for (x, y, w, h) in face:
    cv.circle(img, (x + w // 2, y + h // 2), (w // 2), (0, 255, 255), 2)
```

### Display the image

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)
plt.axis('off')
# plt.savefig("guy2.png", bbox_inches='tight')
```
