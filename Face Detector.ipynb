# My Face Detector App

```python
# Import the OpenCV Package
import cv2

imagePath = 'D:\Videos\Python\My Projects\Images/input_image3.jpg'
```

```python
# Read the image
img = cv2.imread(imagePath)
img.shape
```

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.shape
```

```python
# Load the Classifier
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
# Perform face detection
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=7, minSize=(40, 40)
)
```

```python
# Drawing a bounding box
for (x, y, w, h) in face:
    cv2.circle(img, (x + w // 2, y + h // 2), (w // 2), (0, 255, 255), 2)
```

```python
# Displaying the image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Import the Matplotlib library to display the image
import matplotlib.pyplot as plt
plt.figure(figsize=(40, 40))
plt.imshow(img_rgb)
plt.axis('off')
```
